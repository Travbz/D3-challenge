```python
btc_pearsonr = btc_df.btcopen
eth_pearsonr = eth_df.ethopen
correlation1 = st.pearsonr(btc_pearsonr, eth_pearsonr)
print(f"The correlation between eth/btc is {round(correlation1[0],2)}")
link_pearsonr = link_df.linkopen
corr2 = st.pearsonr(btc_pearsonr, link_pearsonr)
print(f"The correlation between link/btc is {round(corr2[0],2)}")
ada_pearsonr = ada_df.adaopen
corr2 = st.pearsonr(btc_pearsonr, ada_pearsonr)
print(f"The correlation between ada/btc is {round(corr2[0],2)}")
doge_pearsonr = doge_df.dogeopen
corr2 = st.pearsonr(btc_pearsonr, doge_pearsonr)
print(f"The correlation between doge/btc is {round(corr2[0],2)}")
aave_pearsonr = aave_df.aaveopen
corr2 = st.pearsonr(btc_pearsonr, aave_pearsonr)
print(f"The correlation between aave/btc is {round(corr2[0],2)}")
```
